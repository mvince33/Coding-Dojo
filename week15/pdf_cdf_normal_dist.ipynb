```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Load the data
df = pd.read_csv('data/weight-height.csv')
df.head()
```

```python
# Separate the data frame by gender
males = df[df['Gender'] == 'Male'].copy()
females = df[df['Gender'] == 'Female'].copy()
```

```python
# Isolate the variable we are working with
fh = females['Height']
```

```python
low = 66.0
high = 70.0
```

```python
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(x = fh, stat = 'density', kde = True);
ax.set_title('Distribution of Female Heights');
ax.axvline(low, color = 'black');
ax.axvline(high, color = 'black');
ax.axvspan(low, high, color = 'yellow', zorder = 0)
```

```python
mean = np.mean(fh)
std = np.std(fh)
min = np.min(fh)
max = np.max(fh)
```

```python
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(x = fh, stat = 'density', kde = True, cumulative = True);
ax.axvline(mean, color = 'black', ls = '--');
```

```python
x_range = np.linspace(min, max, 1000)
```

```python
pdf = stats.norm.pdf(x_range, loc = mean, scale = std)
```

```python
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(x = fh, stat = 'density');
ax.plot(x_range, pdf);
```

```python
cdf = stats.norm.cdf(x_range, loc = mean, scale = std)
```

```python
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(x = fh, stat = 'density', cumulative = True)
ax.plot(x_range, cdf)
ax.axvline(low, color = 'black');
ax.axvline(high, color = 'black');
ax.axvspan(low, high, color = 'yellow', zorder = 0);
```

```python
stats.norm.cdf(high, loc = mean, scale = std) -\
stats.norm.cdf(low, loc = mean, scale = std)
```
