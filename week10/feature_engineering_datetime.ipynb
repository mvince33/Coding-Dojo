```python
# Imports
import numpy as np
import pandas as pd
```

```python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQgid_gu5oxam-U3pUfWPVIPtOrqs3xOgV9pj9N9JU94AT2mKTVi062N7RMIYHRtuXoGXqgUmVtRqkX/pub?output=csv'
df = pd.read_csv(url, usecols = ['date', 'price', 'bedrooms', 'bathrooms'])
df.head()
```

```python
# Convert the date column to a datetime type
df['date'] = pd.to_datetime((df['date']))
df.info()
```

```python
# Illustrate some of the Pandas methods
# for datetime objects
df['year'] = df['date'].dt.year
df['month (numeric)'] = df['date'].dt.month
df['month (name'] = df['date'].dt.month_name()
df['day of month'] = df['date'].dt.day
df['day of week (numeric)'] = df['date'].dt.weekday
df['day of week (name)']= df['date'].dt.day_name()
df.head()
```
