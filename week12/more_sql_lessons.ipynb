```python
# Imports
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Load the data
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQx3gMRpWbOuReQgWPAKQpMZtqrlrHwOeTUwkWnh8ANvKBi8ItwCBNk5Zrtged021WwgoczTkaqnvAJ/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
# Create a sqlite engine
engine = create_engine('sqlite:///schools.db', echo = True)
# Create a connection to the engine
conn = engine.connect()
```

```python
# Add the data frame as a table
sqlite_table = "school_details"
df.to_sql(sqlite_table, conn, if_exists = 'fail')
```

# Code for the Aggregates lesson

```python
# SQL query string
sql = """
SELECT AVG(`CRCDistance (in Km)`)
FROM school_details;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT MAX(`DistrictHQDistance (in Km)`)
FROM school_details;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT DISTINCT LocatedInRuralOrUrban
FROM school_details;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

# Code for the Aliasing lesson

```python
# SQL query string
sql = """
SELECT EstablishmentName AS col1
FROM school_details;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

# Code for GROUP BY lesson

```python
# SQL query string
sql = """
SELECT LocatedInRuralOrUrban, AVG(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY LocatedInRuralOrUrban;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT PostOfficeName, MAX(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY PostOfficeName;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
ORDER BY COUNT(*) DESC;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT CRCName, COUNT(*) AS NumberOfSchools
FROM school_details
GROUP BY CRCName
ORDER BY NumberOfSchools DESC;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

# Code for HAVING lesson

```python
# SQL query string
sql = """
SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
HAVING COUNT(*) > 20;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT CRCName, COUNT(*) AS NumberOfSchools
FROM school_details
GROUP BY CRCName
HAVING NumberOfSchools > 20
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT SubdivisionName, AVG(`DistrictHQDistance (in Km)`) AS Avg_HQ_Distance
FROM school_details
GROUP BY SubdivisionName
HAVING Avg_HQ_Distance > 30;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```
