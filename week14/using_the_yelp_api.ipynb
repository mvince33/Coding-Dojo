```python
# Imports
import numpy as np
import pandas as pd
import json
from yelpapi import YelpAPI
```

```python
# Get the credentials
with open('yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# Instantiate a YelpAPI object
yelp_api = YelpAPI(login['api-key'], timeout_s = 5.0)
```

```python
# Make a query
search_results = yelp_api.search_query(location = 'Seattle, WA',
                                       term = 'Vegetarian')
```

```python
# Find the number of results
search_results['total']
```

```python
# Save the first 20 results as a data frame
biz = pd.DataFrame(search_results['businesses'])
biz.head()
```

```python
# Get the next 20 results
search_results = yelp_api.search_query(location = 'Seattle, WA',
                                       term = 'Vegetarian',
                                       offset = 20)
```

```python
# Save those results as a data frame
biz20 = pd.DataFrame(search_results['businesses'])
biz20.head()
```

```python
# Concatanate the results into one data frame
businesses = pd.concat([biz, biz20], ignore_index = True)
display(businesses.head(), businesses.tail())
```

```python
# Save the results as a JSON file
records = businesses.to_json(orient = 'records')

with open('yelp_records.json', 'w') as f:
    json.dump(records, f)
```
