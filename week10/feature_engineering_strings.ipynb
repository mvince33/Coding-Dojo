```python
# Imports
import numpy as np
import pandas as pd
```

```python
# Load the data
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTduFDqlX0txyWdDgpOvIE7XsAT0tINDXSO4pMa_aj9GbnDdXVwPLHOc3kMv9lVXUeqXBCl6gp9Bj2Q/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
# Split the name column into two 
# columns FirstName and LastName
df[['LastName', 'FirstName']] = df['Name'].str.split(',', expand = True)
# Drop the name column
df.drop('Name', axis = 1, inplace = True)
df.head()
```

```python
df.loc[0, 'FirstName']
```

```python
# Strip the extra whitespace from 
# the LastName column
df['FirstName'] = df['FirstName'].str.strip()
df.loc[0, 'FirstName']
```

```python
# Recombine FirstName and LastName into one column.
df['Name'] = df['FirstName'] + ' ' + df['LastName']
df.drop(columns = ['FirstName', 'LastName'], inplace = True)
df.head()
```
