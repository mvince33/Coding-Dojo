```python
# Imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
# Load the data
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQ8RFVjiTdVKNZfDw_wbgzrI9MjcqU3U_jJXQjeEA1bKqJZu8x-FDKAwnilzeauvEjQZOvpZR4ACA0s/pub?output=csv'
wine = pd.read_csv(url)
wine.head()
```

```python
# Get the features we are focusing on
df = wine[['malic_acid', 'flavanoids']]
df.head()
```

```python
# Scale the data
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
# Plot the dendrogram
plt.figure(figsize = (15, 5));
sch.dendrogram(sch.linkage(scaled_df, method = 'ward'));
plt.xlabel('Data Points');
```
