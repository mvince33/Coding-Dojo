```python
# Imports
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Load the data
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQx3gMRpWbOuReQgWPAKQpMZtqrlrHwOeTUwkWnh8ANvKBi8ItwCBNk5Zrtged021WwgoczTkaqnvAJ/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
# Create a sqlite engine
engine = create_engine('sqlite:///schools.db', echo = True)
# Create a connection to the engine
conn = engine.connect()
```

```python
# Add the dataframe as a table
sqlite_table = "school_details"
df.to_sql(sqlite_table, conn, if_exists = 'fail')
```

```python
# SQL query string
sql = """
SELECT EstablishmentName, SchoolLevel, Website
FROM school_details
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT EstablishmentName, `DistrictHQDistance (in Km)`
FROM school_details
WHERE `DistrictHQDistance (in Km)` > 100;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
FROM school_details
WHERE SchoolLevel = 'JHS';
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
FROM school_details
WHERE EstablishmentName LIKE '%JUNIOR HIGH%'
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```

```python
# SQL query string
sql = """
SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
FROM school_details
WHERE EstablishmentName LIKE 'C%'
ORDER BY EstablishmentName
LIMIT 10;
"""

# Read the sql query
pd.read_sql_query(sql, engine)
```
