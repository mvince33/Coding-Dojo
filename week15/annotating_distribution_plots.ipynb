```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Load the data
df = pd.read_csv('data/medical_data.csv', index_col = 0)
df.head()
```

```python
# Plot the histogram for income
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(df['Income'], ax = ax, stat = 'probability');
```

```python
# Get the mean
mean = df['Income'].mean()
mean
```

```python
# Plot the histogram for income with vertical line for the mean
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(df['Income'], ax = ax, stat = 'probability');
ax.axvline(mean, color = 'red', label = 'Mean');
ax.set_title('Income');
ax.legend();
```

```python
# Make the above reproducible
col = 'Income'
feature = df[col]
mean = feature.mean()
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(feature, ax = ax, stat = 'probability');
ax.axvline(mean, color = 'red', label = f'Mean = {mean: .2f}');
ax.set_title(f'{col}');
ax.legend();
```

```python
col = 'Initial_days'
```

```python
feature = df[col]
mean = feature.mean()
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(feature, ax = ax, stat = 'probability');
ax.axvline(mean, color = 'red', label = f'Mean = {mean: .2f}');
ax.set_title(f'{col}');
ax.legend();
```

```python
# Add the median to the Income histogram
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(feature, ax = ax, stat = 'probability');
ax.axvline(mean, color = 'red', label = f'Mean = {mean: .2f}');
ax.axvline(median, color = 'green', ls = '--', 
           label = f'Median = {median: 0.2f}')
ax.set_title(f'{col}');
ax.legend();
```

```python
# Highlight the region between 1 standard deviation from the mean
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
po_std = mean + std
mo_std = mean - std
fig, ax = plt.subplots(figsize = (13, 8));
sns.histplot(feature, ax = ax, stat = 'probability');
ax.axvline(mean, color = 'red', label = f'Mean = {mean: .2f}');
ax.axvline(median, color = 'green', ls = '--', 
           label = f'Median = {median: 0.2f}');
ax.axvline(po_std, color = 'black', label = f'+1 std: {po_std: 0.2f}');
ax.axvline(mo_std, color = 'black', label = f'-1 std: {mo_std: 0.2f}');
ax.axvspan(mo_std, po_std, color = 'yellow', zorder = 0);
ax.set_title(f'{col}');
ax.legend();
```
