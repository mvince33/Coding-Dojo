```python
# Imports
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Load the data
school = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQx3gMRpWbOuReQgWPAKQpMZtqrlrHwOeTUwkWnh8ANvKBi8ItwCBNk5Zrtged021WwgoczTkaqnvAJ/pub?output=csv')
facility = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRki7gKY1pRqDvrratwkMWJA0ly4Z390oEoeS0EaDhcNmzSfXqxD13R_Fj7AnJKnDoghODI8SVSNS1L/pub?output=csv')
display(school.head())
facility.head()
```

```python
# Create sqlite engine
engine = create_engine('sqlite:///school.db', echo = True)
# Create connection to engine
conn = engine.connect()
```

```python
# Add the data frames as tables in sqlite
sqlite_table1 = "school_details"
school.to_sql(sqlite_table1, conn, if_exists = 'fail')
sqlite_table2 = "school_facility_details"
facility.to_sql(sqlite_table2, conn, if_exists = 'fail')
```

```python
# SQL query
sql = """
SELECT AVG (`BACDistance (in Km)`) as avg_bac_distance
FROM school_details
"""

# Read the sql query into pandas
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# SQL query
sql = """
SELECT SchoolLevel, AVG (`BACDistance (in Km)`) as avg_bac_distance
FROM school_details
GROUP BY SchoolLevel
ORDER BY avg_bac_distance DESC;
"""

# Read the sql query into pandas
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# SQL query
sql = """
SELECT SchoolLevel, AVG (`BACDistance (in Km)`) as avg_bac_distance
FROM school_details
GROUP BY SchoolLevel
HAVING avg_bac_distance >= 10
ORDER BY avg_bac_distance DESC;
"""

# Read the sql query into pandas
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# SQL query
sql = """
SELECT *
FROM school_details sd 
INNER JOIN school_facility_details sfd
ON sd.EstablishmentName = sfd.EstablishmentName
LIMIT 2;
"""

# Read the sql query into pandas
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# SQL query
sql = """
SELECT *
FROM school_details sd 
INNER JOIN school_facility_details sfd
ON sd.EstablishmentName = sfd.EstablishmentName
WHERE sd.SchoolLevel = 'PS' AND sfd.IsLibraryAvailable != 'Available';
"""

# Read the sql query into pandas
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# SQL query
sql = """
SELECT *
FROM school_details sd 
INNER JOIN school_facility_details sfd
ON sd.EstablishmentName = sfd.EstablishmentName
WHERE sd.SchoolLevel = 'PS' AND sfd.IsLibraryAvailable != 'Available'
ORDER BY sd.EstablishmentName;
"""

# Read the sql query into pandas
df = pd.read_sql_query(sql, engine)
df.head()
```
