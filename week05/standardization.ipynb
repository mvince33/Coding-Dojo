```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQ64T8zJ-HDpXMNwvvCaW70p1BtIVmPg9D0JKVSFXp7f6eDdFbB_nT12TwVipUMUtBbAlfS4S5MjCza/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
# Set the target
y = df['Sold']
# Set the features
X = df.drop(columns = 'Sold')
```

```python
# Train, test, split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Get descriptive stats on test set. 
X_train.describe().round(2)
```

```python
# Instantiate the scalar
scaler = StandardScaler()
scaler.fit(X_train)
X_train.describe()
```

```python
# Scale the training and testing data
train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)
train_scaled[:10]
```

```python
# Transform the training set back to a DataFrame
X_train_scaled = pd.DataFrame(train_scaled, columns = X_train.columns)
X_train_scaled
```

```python
# Get descripctive stats
X_train_scaled.describe().round(2)
```
