```python
from google.colab import drive
import numpy as np
import pandas as pd
path = '/content/drive/MyDrive/Coding Dojo/02 Week 2: Pandas/RailsToTrails_National Count Data_week.xlsx'
bike_df = pd.read_excel(path)
```

```python
bike_df.head()
```

```python
bike_df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'], inplace=True)
bike_df.rename(columns={'Unnamed: 6': 'percent_change'}, inplace=True)
bike_df.head()
```

```python
bike_df.isna().sum()
missing_counts = bike_df['2021 Counts'].isna().sum()
print(f"There are {missing_counts} missing values in the dataset.")
```

```python
bike_df['2021 Counts'].fillna(0, inplace=True)
```

```python
# The only thing I could notice with the name of the 2019 counts (31 counters) column is that
# there was some whitespace at the beginning of the name.
bike_df.rename(columns={' 2019 counts (31 counters)': 'counts_2019'}, inplace=True)
```

```python
# Better fix the other one while were at it!
bike_df.rename(columns={' 2020 counts (31 counters)': 'counts_2020'}, inplace=True)
```
